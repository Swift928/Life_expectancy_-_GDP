# Introduction
The goal of this project is to explore and analyze data collected from the World Health Organization and the World Bank, to reveal any patterns in life expectancy and gdp of a country. In the data provided, six countries were chosen. Below are the questions this project will answer.



Questions
- What are the average life expectancy of the countries?
- What country had the highest increase in life expectancy?
- What country had the highest increase in gdp?
- Is there a positive correlation between life expectancy and gdp?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a new datafrom with the Life expectancy Gdp data

df = pd.read_csv('all_data.csv')
```

```python
# Display a sample of 10 rows

display(df.sample(10))
```

```python
# Check for any duplicate rows

df.duplicated().sum()
```

```python
# Check for missing data

df.isna().sum()
```

```python
# Display the column datatypes

display(df.dtypes)
```

The data types here are what we expect and don't require any change.

```python
# Change column names to lower case
df.columns = [x.lower() for x in df.columns]

# Rename life expectancy column name for ease of use
df.rename(columns ={'life expectancy at birth (years)': 'life_expectancy'}, inplace=True)

# Rename United States of America to U.S.A

df['country'].replace('United States of America', 'USA', inplace=True)

#  Check that it worked and also that the data looks as expected(No missing values)
df.info()

# Make every gdp of 1 billion
df.gdp = df.gdp/1000000000
```

```python
# Display unique countries and year 

country = df.country.unique().tolist()
print(country)

year = df.year.unique().tolist()
print(year)
```

```python
# Find average life expectancy of each country

average_life_expectancy = df.groupby('country', as_index=False).life_expectancy.mean().sort_values(by='life_expectancy', ascending=False).round(2).reset_index().drop(columns='index')


display(average_life_expectancy)
```

```python
iqr_range = {}

for x in country:
    data = df[df['country'] == x]
    
    q1, q3 = np.percentile(data.life_expectancy,[25, 75])
    
    iqr_range[x] = round(q3 - q1, 3) 
    
display(iqr_range)
```

```python
# More life expectancy analysis

plt.figure(figsize=(12,8))

plt.subplot(1,2,1)

sns.boxplot(data=df, x='country', y='life_expectancy')

plt.xlabel('')
plt.ylabel('Years')
plt.title('Life Expectancy')


plt.subplot(1,2,2)


values = list(iqr_range.values())
names = list(iqr_range.keys())

plt.bar(range(len(iqr_range)), values, tick_label=names)

plt.title('Interquartile ranges')

for x in range(len(iqr_range)):
    plt.text(x - .15, values[x] + .05, values[x])
    

plt.tight_layout()
plt.show()
```

From these visuals of life expactancy, we can see that Zimbabwe is much lower on the life expectancy scale. The life expectancy is roughly 25 years lower than second lowest country, China. This is a major difference and within the next analysis I plan to discover if there is a correlation between life expectancy and gdp.

In the second visual, Zimbabwe leads the others for interquartile range of life expectancy. This would be interesting to notice if its GDP also grew inline with such life expectancy improvement.

```python
# Create a year over year comparison of gdp by country

plt.figure(figsize=(15,7))

sns.pointplot(data=df, x='year', y='gdp', hue='country')

plt.ylabel('GDP: Billions USD', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.legend(title='Country')


plt.suptitle('GDP: Year over Year')
plt.tight_layout()
plt.show()
```

From this visual we notice the USA having the highest GDP by nearly 5-7 trillion year over year. I would like to see if this also correlates to a higher life expectancy. We also see an interesting over-lap with China and Germany. For the first few years of recording Germany is leading China until about 2007 when China out-paces Germany towards closing the gap with the USA. This would be interesting to analyze if life expectancy followed a similar trend for China.

Lastly, the GDP of Chile is interestingly low given it's high life expectancy, second among the other countries behind Germay.

```python
# Create a FacetGrid displaying GDP

g = sns.FacetGrid(df, col='country', hue='country', col_wrap=3, sharey=False, aspect=1.5)
g.map_dataframe(sns.pointplot, x='year', y='gdp')
g.set_axis_labels('Year' ,'GDP: Billions USD')

plt.xticks(np.arange(0, len(year), 2))


plt.show()
```

In this second visual on GDP, we can see each countries GDP growth over the years on individual scales. From this view we can see the growth in GDP for countries like Zimbabwe and Chile, whose GDP seems stagnant compared to the other countries. 

This visual can be used to explain the major interquartile range for Zimbabwe, whose GDP doubled during this recording. Eventhough Zimbabwe is not the only country to increase GDP nearly 100%, it does have the lowest GDP of the other countries. From this, I will hypothesize that after a certain GDP has been achieved, increases in life expectancy start to narrow/diminish.

```python
# Create a year over year graph of life expectancy by country

plt.figure(figsize=(10, 8))
sns.pointplot(data=df, x='year', y='life_expectancy', hue='country')
plt.ylabel('Life Expectancy' , fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.title('Life Expectancy: Year over Year')
plt.show()
```

In this graph we can see the impressive rise in life expectancy for Zimbabwe.

```python
# Create a FacetGrid displaying life expectancy

g = sns.FacetGrid(df, col='country', col_wrap=3, hue='country', sharey=False, aspect=1.5)
g.map_dataframe(sns.pointplot, x='year', y='life_expectancy')
g.set_axis_labels('Year', 'Life Expectancy')

plt.xticks(np.arange(0, len(year), 2))

plt.show()
```

This visual made it easier to notice the change in life expectancy for the other countries that moved less, in comparison to Zimbabwe.

```python
# Create a custom pointplot for each country displaying gdp and life expectancy on seperate axes

fig = plt.figure(figsize=(15,8))

for numb, name in enumerate(country, 1):
    

    ax1 = plt.subplot(2,3,numb)
    
    sns.pointplot(x=df[df.country == name].year, y=df[df.country == name].gdp, marker='o', color='red')
    plt.ylabel('GDP: Billions USD', color='red')
    plt.xlabel('Year', fontsize=10)
    ax1.set_xlabel('')
    

    ax2 = ax1.twinx()
    
    sns.pointplot(x=df[df.country == name].year, y=df[df.country == name].life_expectancy, marker='o', color='green')
    ax2.set_ylabel('Life Expectancy', color='green')
    plt.title(name)
    
    
    
    if numb in([1,2,3]):
        plt.xticks(np.arange(0, len(year), 2))
        ax1.set_xticklabels('')
    else:
        plt.xticks(np.arange(0, len(year), 2))
    
    if numb in([2,3,5,6]):
        ax1.set_ylabel('')
    else:
        pass
    
    if numb in([1,2,4,5]):
        ax2.set_ylabel('')
    else:
        pass


plt.tight_layout(pad=1.5)
plt.show()
```

This visual is unique, in that each graph displays seperate y-axis calculations. On the left y-axis, the gdp is recorded, and on the right y-axis, life expectancy is recorded. This visual makes changes in gdp and life expectancy easy to track over time. For example, the dip in gdp for Zimbabwe can be traced easily along with it's slight drop in life expectancy.

```python
g = sns.FacetGrid(data=df, col='country', hue='country', aspect=1.5, sharex=False, sharey=False, col_wrap=3)
g.map_dataframe(sns.scatterplot, x='life_expectancy', y='gdp')
g.set_axis_labels('Life Expectancy', 'GDP: Billions USD')

plt.tight_layout()
plt.show()
```

Similar to the previous visual, this visual measures life expectancy and gdp, not over time but how one moved vs the other.
